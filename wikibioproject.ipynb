```python
#imput about the name of wiki
import requests
import bs4
from PIL import Image
from io import BytesIO
from IPython.display import display
#name is printed

#info is printed

#photo link is given
link='https://en.wikipedia.org/wiki/{}'
while True:
      choice = input('Do you want to continue y or n: ')
      if choice.lower().startswith("y"):
         
            about=input('who do you want a info of: ')
            fi_url=link.format(about)
        
            re=requests.get(fi_url)
            soup=bs4.BeautifulSoup(re.text,'lxml')

            name=soup.select('.mw-page-title-main')[0].text
            labels=soup.select('.infobox-label')
            data_entries=soup.select('.infobox-data')
            for label,data in zip(labels,data_entries):
                print(f'{label.text.strip()}:\n{data.text.strip()}\n')#strip cleans it out

            #image
            try:
                image_url="https:"+ soup.select('.infobox-image a img')[0]['src']
                response=requests.get(image_url)
                image = Image.open(BytesIO(response.content))
                print(f'\n\nPicture of {name}: ')
                display(image)
            except:
                print('No image found.')
                
      elif choice.lower().startswith("n"):
        print("Exiting the loop.")
        break
      else:
        print("Invalid input. Please enter 'y' or 'n'.")
```
